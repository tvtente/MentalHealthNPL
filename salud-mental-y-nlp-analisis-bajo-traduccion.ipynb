  <h3 style="margin:0;color:#000000;font-size:22px;">1. Modelo utilizado e infraestructura de ejecución</h3>
  <div style="font-size:78%;font-weight:normal;color:#1D3037;margin-top:5px;">

  </div>

<div style="margin:16px 0 20px 0;padding:16px 20px;border-left:6px solid #C0DDD1;background:#FFFFFF;border-radius:14px;font-family:Arial,Helvetica,sans-serif;color:#1D3037;box-shadow:0 4px 14px rgba(0,0,0,0.06);">
  <h3 style="margin:0;color:#000000;font-size:19px;">¿Qué modelo se utilizó y cómo fue puesto en funcionamiento?</h3>
  <p style="font-size:15px;line-height:1.65;margin:10px 0 0;">
    La traducción fue ejecutada mediante el modelo <b>gemma3:4b</b>, desplegado localmente con <b>Ollama</b>. El modelo puede descargarse desde la biblioteca oficial de Ollama y ejecutarse mediante su servicio local, lo que permite exponer un endpoint de inferencia dentro de la propia máquina de trabajo. Esta arquitectura hizo posible controlar el flujo desde scripts personalizados de Python, sin depender de llamadas continuas a servicios externos de traducción.
  </p>
  <p style="font-size:15px;line-height:1.65;margin:10px 0 0;">
    La elección de inferencia local respondió a criterios de coste, reproducibilidad, continuidad operativa y flexibilidad metodológica. En lugar de limitar el proceso a un traductor generalista o a una API comercial, se optó por un entorno que permitiera ajustar prompts, tamaños de lote, tiempos de espera, validación de formato y estrategias de recuperación ante fallos conforme avanzaba la complejidad del corpus.
  </p>
  <p style="font-size:15px;line-height:1.65;margin:10px 0 0;">
    El procedimiento implementado no se limita exclusivamente a la traducción hacia español. Dado que el flujo permite parametrizar tanto el idioma de origen como el idioma o variante lingüística de destino mediante códigos de locale, es posible generar versiones del corpus adaptadas a distintos idiomas soportados y, dentro de ellos, a diferentes regiones de uso, por ejemplo <b>es-ES</b>, <b>es-AR</b>, <b>en-US</b> o <b>en-GB</b>. Esta flexibilidad resulta metodológicamente relevante, ya que un modelo orientado a clasificación o análisis lingüístico puede requerir alineación no solo con una lengua general, sino también con convenciones léxicas, pragmáticas y culturales propias de la comunidad en la que se desea evaluar su rendimiento.
  </p>
  <p style="font-size:15px;line-height:1.65;margin:10px 0 0;">
    En este sentido, la traducción por locale no debe entenderse únicamente como una diferencia ortográfica o regional, sino como una estrategia de adaptación lingüística del corpus a contextos de uso específicos. Esto permite construir datasets derivados más coherentes con la variedad idiomática del entorno de aplicación, reduciendo el riesgo de introducir formas poco naturales o culturalmente desalineadas en fases posteriores de análisis, entrenamiento o evaluación.
  </p>

<div style="text-align:center;border-radius:24px;padding:12px;margin:20px 0 14px 0;font-size:108%;font-family:Arial,Helvetica,sans-serif;background:#F4D35E;color:#1D3037;overflow:hidden;border:1px solid rgba(29,48,55,0.25);box-shadow:0 4px 14px rgba(0,0,0,0.12);">
  <h3 style="margin:0;color:#000000;font-size:22px;">2. Estructura operativa del flujo de traducción</h3>
  <div style="font-size:78%;font-weight:normal;color:#1D3037;margin-top:5px;">
    Organización del archivo fuente, archivo de trabajo y criterio de continuidad sobre las celdas traducidas
  </div>
</div>

<div style="margin:16px 0 20px 0;padding:16px 20px;border-left:6px solid #F4D35E;background:#F8FAF8;border-radius:14px;font-family:Arial,Helvetica,sans-serif;color:#1D3037;box-shadow:0 4px 14px rgba(0,0,0,0.06);">
  <h3 style="margin:0;color:#000000;font-size:19px;">¿Cómo se organizó el procesamiento para no perder avance?</h3>
  <p style="font-size:15px;line-height:1.65;margin:10px 0 0;">
    El flujo principal se apoyó en un archivo fuente con el texto original y en un archivo de trabajo en formato CSV que acumuló progresivamente las traducciones en español. El sistema añadió una columna específica para la salida traducida y utilizó esa columna como criterio real de continuación: las filas con contenido se consideraron resueltas y las filas vacías se consideraron pendientes.
  </p>
  <p style="font-size:15px;line-height:1.65;margin:10px 0 0;">
    Este patrón hizo que la reanudación no dependiera del número absoluto de línea procesada, sino del estado efectivo del dataset de trabajo. Si la ejecución se interrumpía, bastaba con relanzar el proceso para que el sistema reconstruyera la lista de pendientes a partir de los espacios todavía vacíos en la columna traducida, sin reiniciar desde cero.
  </p>
  <p style="font-size:15px;line-height:1.65;margin:10px 0 0;">
    Para el caso de trabajo desarrollado en este proyecto, la ejecución controlada puede realizarse indicando explícitamente el archivo fuente, el archivo de trabajo, el locale de origen y destino, el tamaño de lote, la longitud de fragmento y el timeout. Un ejemplo representativo sería una configuración como <b>source-locale = en-US</b>, <b>target-locale = es-ES</b>, <b>batch-size = 8</b>, <b>chunk-max-chars = 500</b> y <b>request-timeout = 120</b>, lo que permite mantener una traducción gradual, validada y reanudable sobre el corpus actual.
  </p>
</div>

<a id="batching-strategy"></a>
<div style="text-align:center;border-radius:24px;padding:12px;margin:20px 0 14px 0;font-size:108%;font-family:Arial,Helvetica,sans-serif;background:#F4D35E;color:#1D3037;overflow:hidden;border:1px solid rgba(29,48,55,0.25);box-shadow:0 4px 14px rgba(0,0,0,0.12);">
  <h3 style="margin:0;color:#000000;font-size:22px;">3. Segmentación por lotes y ajuste dinámico</h3>
  <div style="font-size:78%;font-weight:normal;color:#1D3037;margin-top:5px;">
    Uso de lotes pequeños, reducción automática y adaptación al incremento de textos largos y complejos
  </div>
</div>

<div style="margin:16px 0 20px 0;padding:16px 20px;border-left:6px solid #C0DDD1;background:#FFFFFF;border-radius:14px;font-family:Arial,Helvetica,sans-serif;color:#1D3037;box-shadow:0 4px 14px rgba(0,0,0,0.06);">
  <h3 style="margin:0;color:#000000;font-size:19px;">¿Por qué fue necesario trabajar con lotes pequeños y variables?</h3>
  <p style="font-size:15px;line-height:1.65;margin:10px 0 0;">
    A medida que el corpus avanzó, las filas más sencillas tendieron a resolverse primero y quedaron pendientes textos progresivamente más largos, narrativos y difíciles de serializar correctamente en la salida. Por esta razón, el tamaño de lote fue reducido hasta una configuración más estable de <b>8 filas por solicitud</b>, especialmente adecuada para esta fase avanzada del dataset.
  </p>
  <p style="font-size:15px;line-height:1.65;margin:10px 0 0;">
    Además, el sistema fue diseñado para reducir automáticamente el lote cuando aparecían errores de formato, truncamiento, validación o timeout. La lógica aplicada fue una reducción por mitades, de modo que un lote podía pasar de <b>8 a 4</b>, luego a <b>2</b> y finalmente a <b>1</b>. Esto permitió distinguir entre fallos causados por el volumen conjunto del lote y fallos debidos a la longitud o complejidad de una sola fila.
  </p>
</div>

  <div style="font-size:78%;font-weight:normal;color:#1D3037;margin-top:5px;">

  <h3 style="margin:0;color:#000000;font-size:19px;">¿Cómo se evitó incorporar respuestas dañadas al dataset?</h3>
  <p style="font-size:15px;line-height:1.65;margin:10px 0 0;">
    El modelo fue instruido para devolver la salida en formato CSV con pares <b>row_id, translation_es</b>. Sin embargo, en la práctica se observaron respuestas truncadas, filas omitidas, identificadores equivocados y salidas semánticamente correctas pero estructuralmente inválidas. Para impedir que estas respuestas dañaran el dataset, se incorporó una capa de validación estricta antes de escribir cualquier resultado en el archivo de trabajo.
  </p>
  <p style="font-size:15px;line-height:1.65;margin:10px 0 0;">
    La validación comprobó que la respuesta incluyera exactamente los identificadores esperados, sin omisiones, duplicados ni row_ids ajenos al lote solicitado. Cuando esta condición no se cumplía, la salida no se integraba al archivo final, sino que el sistema la trataba como error de lote y reintentaba con un tamaño menor, preservando la integridad tabular del corpus traducido.
  </p>
</div>

  <h3 style="margin:0;color:#000000;font-size:22px;">5. Gestión de bloqueos y tiempos de espera</h3>
  <div style="font-size:78%;font-weight:normal;color:#1D3037;margin-top:5px;">
    Incorporación de timeout por petición y tratamiento de respuestas lentas como parte del circuito de recuperación
  </div>
</div>

<div style="margin:16px 0 20px 0;padding:16px 20px;border-left:6px solid #C0DDD1;background:#FFFFFF;border-radius:14px;font-family:Arial,Helvetica,sans-serif;color:#1D3037;box-shadow:0 4px 14px rgba(0,0,0,0.06);">
  <h3 style="margin:0;color:#000000;font-size:19px;">¿Qué se hizo cuando el modelo se bloqueaba o tardaba demasiado?</h3>
  <p style="font-size:15px;line-height:1.65;margin:10px 0 0;">
    Durante la ejecución se detectaron situaciones donde el modelo local no fallaba explícitamente, pero quedaba excesivamente lento o aparentemente congelado sobre determinados textos. Para evitar que todo el proceso quedara suspendido indefinidamente, se introdujo un <b>timeout por petición</b>, de manera que cualquier solicitud que superara el tiempo configurado fuera tratada como un fallo controlado.
  </p>
  <p style="font-size:15px;line-height:1.65;margin:10px 0 0;">
    Esta decisión permitió incorporar los bloqueos dentro del mismo circuito de recuperación que los errores de formato o truncamiento. En términos operativos, un timeout no implicó abortar toda la traducción, sino reducir el lote o, si el problema persistía en tamaño mínimo, derivar la fila a una estrategia específica de rescate.
  </p>
</div>

  <div style="font-size:78%;font-weight:normal;color:#1D3037;margin-top:5px;">
    Registro auxiliar de casos conflictivos y separación entre fallos de lote y fallos de longitud individual

<div style="margin:16px 0 20px 0;padding:16px 20px;border-left:6px solid #F4D35E;background:#F8FAF8;border-radius:14px;font-family:Arial,Helvetica,sans-serif;color:#1D3037;box-shadow:0 4px 14px rgba(0,0,0,0.06);">
  <h3 style="margin:0;color:#000000;font-size:19px;">¿Qué ocurrió con las filas que seguían fallando incluso en lote mínimo?</h3>
  <p style="font-size:15px;line-height:1.65;margin:10px 0 0;">
    Cuando una fila continuaba fallando incluso con lote unitario, el problema dejó de interpretarse como un error de agrupación y pasó a considerarse un problema de longitud o complejidad individual de la celda. Para estos casos se creó un archivo auxiliar de problemáticas, donde se registraron el identificador de fila, la etiqueta contextual, la longitud del texto, el motivo del fallo, el texto fuente y la fecha de registro.
  </p>
  <p style="font-size:15px;line-height:1.65;margin:10px 0 0;">
    Esta separación permitió que el flujo principal siguiera avanzando sin perder trazabilidad sobre los casos difíciles. En lugar de bloquear toda la ejecución por unas pocas narraciones extensas o conflictivas, cada excepción quedó documentada y disponible para una intervención posterior o inmediata mediante otro mecanismo de rescate.
  </p>
</div>

  <h3 style="margin:0;color:#000000;font-size:19px;">¿Cómo se rescataron los textos demasiado largos para una sola respuesta?</h3>
  <p style="font-size:15px;line-height:1.65;margin:10px 0 0;">
    Para las filas que seguían provocando truncamiento incluso en tamaño unitario, se implementó una estrategia de <b>traducción fragmentada</b>. El texto original se dividió en fragmentos de aproximadamente <b>500 caracteres</b>, procurando respetar cortes razonables por frases y unidades discursivas. Cada fragmento se envió al modelo por separado y las traducciones parciales se recombinaron posteriormente en una sola salida final.
  </p>
  <p style="font-size:15px;line-height:1.65;margin:10px 0 0;">
    Esta recuperación por fragmentos fue integrada dentro del propio flujo principal. Es decir, una fila podía ser detectada como problemática, registrada en el archivo auxiliar y, a continuación, sometida de inmediato a un intento de rescate. Si la recomposición resultaba exitosa, la traducción se escribía en el archivo de trabajo y la fila se eliminaba de la cola de problemáticas; si no, quedaba pendiente para una segunda pasada específica sobre textos largos.

## 2.2. Estructura textual considerada

````python
# =========================
# 1. Definir texto disponible para caracterización
# =========================

df_reading = df.copy()

# Texto original del dataset
df_reading['statement_original'] = df_reading['statement'].astype(str).str.strip()

# Detectar columna de traducción si existe
translation_column = None

possible_translation_columns = [
    'translate',
    'translation',
    'statement_es',
    'statement_translate',
    'statement_spanish'
]

for col in possible_translation_columns:
    if col in df_reading.columns:
        translation_column = col
        break

# Crear columna de traducción
if translation_column is not None:
    df_reading['statement_translate'] = df_reading[translation_column].astype(str).str.strip()
else:
    df_reading['statement_translate'] = ''

# Saber si existe traducción usable
df_reading['is_translated'] = (
    df_reading['statement_translate'].notna() &
    (df_reading['statement_translate'].str.strip() != '') &
    (df_reading['statement_translate'].str.lower().str.strip() != 'nan')
)

# Texto disponible para caracterización inicial
# Si hay traducción, se usa español; si no, se usa el original
df_reading['statement_working'] = df_reading['statement_translate'].where(
    df_reading['is_translated'],
    df_reading['statement_original']
)

# Idioma del texto que se está usando para la lectura inicial
df_reading['working_language'] = df_reading['is_translated'].map({
    True: 'es',
    False: 'en'
})

print("Columna de traducción detectada:", translation_column)
print("Filas con traducción:", df_reading['is_translated'].sum())
print("Filas sin traducción:", (~df_reading['is_translated']).sum())
````

````python
# Caracteres en texto original inglés
df_reading['num_of_characters_en'] = df_reading['statement_original'].str.len()

# Frases en texto original inglés
df_reading['num_of_sentences_en'] = df_reading['statement_original'].apply(
    lambda x: len(nltk.sent_tokenize(str(x), language='english'))
)

# Caracteres en texto traducido español
df_reading['num_of_characters_es'] = df_reading['statement_translate'].apply(
    lambda x: len(str(x)) if str(x).strip().lower() != 'nan' and str(x).strip() != '' else 0
)

# Frases en texto traducido español
df_reading['num_of_sentences_es'] = df_reading['statement_translate'].apply(
    lambda x: len(nltk.sent_tokenize(str(x), language='spanish')) 
    if str(x).strip().lower() != 'nan' and str(x).strip() != '' else 0
)

# Caracteres del texto efectivo de trabajo
df_reading['num_of_characters_working'] = df_reading['statement_working'].str.len()

# Frases del texto efectivo de trabajo
df_reading['num_of_sentences_working'] = df_reading.apply(
    lambda row: len(
        nltk.sent_tokenize(
            str(row['statement_working']),
            language='spanish' if row['working_language'] == 'es' else 'english'
        )
    ),
    axis=1
)
````

## 3.2. Removing URLs, handles, punctuation and special characters

To remove specific patterns such as URLs or other unwanted text (like ```[View Poll](https://www.reddit.com/poll/...)```) from a column in a pandas DataFrame, we can use regular expressions with the re module or pandas built-in string methods.

````python
def remove_patterns(text):
    # Remove URLs
    text = re.sub(r'http[s]?://\S+', '', text)
    # Remove markdown-style links
    text = re.sub(r'\[.*?\]\(.*?\)', '', text)
    # Remove handles (that start with '@')
    text = re.sub(r'@\w+', '', text)
    # Remove punctuation and other special characters
    text = re.sub(r'[^\w\s]', '', text)
    return text.strip()

# Apply the function to the 'statement' column
df['statement'] = df['statement'].apply(remove_patterns)
df.head()
````

## 3.3. Tokenization

Split text into individual words or tokens.

````python
# Apply word_tokenize to each element in the 'statement' column
df['tokens'] = df['statement'].apply(word_tokenize)
df.head()
````

## 3.4. Stemming

Reduce words to their base or root form. However, in some cases, the stemming process produces words that are not correct spellings of the root word. For example, **happi** and **sunni**. That's because it chooses the most common stem for related words. For example, we can look at the set of words that comprises the different forms of happy:

- happy
- happiness
- happier

We can see that the prefix **happi** is more commonly used. We cannot choose **happ** because it is the stem of unrelated words like **happen**.

````python
# Initialize the stemmer
stemmer = PorterStemmer()

# Function to stem tokens and convert them to strings
def stem_tokens(tokens):
    return ' '.join(stemmer.stem(str(token)) for token in tokens)

# Apply the function to the 'tokens' column
df['tokens_stemmed'] = df['tokens'].apply(stem_tokens)

df.head()
````

## 3.5. Stop Words

While stop words are typically removed to reduce noise, in sentiment analysis for mental health, many stop words are actually crucial for understanding context and sentiment. Words like negations and those related to emotions (e.g., 'how', 'why', 'because') provide important insights. Removing these can decrease the accuracy of the analysis.

## 4.1. Separate features and labels

````python
X = df[['tokens_stemmed', 'num_of_characters', 'num_of_sentences']]
y = df['status']
````

## 4.2. Label encoding target variable

````python
lbl_enc = LabelEncoder()
y = lbl_enc.fit_transform(y.values)
````

## 4.3. Split the data into training and testing sets

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
````

## 4.4. Convert text to features using TF-IDF vectoriser

Now, we will transform tokens (words) into numerical values that represent the importance of words in a document relative to a collection of documents. This helps highlight unique words in a document while downplaying common ones, making it easier for machine learning models to identify relevant patterns and make better predictions.

````python
# 1. Initialize TF-IDF Vectorizer and fit/transform on the 'tokens' column
vectorizer = TfidfVectorizer(ngram_range=(1, 2), max_features=50000)
X_train_tfidf = vectorizer.fit_transform(X_train['tokens_stemmed'])
X_test_tfidf = vectorizer.transform(X_test['tokens_stemmed'])

# 2. Extract numerical features
X_train_num = X_train[['num_of_characters', 'num_of_sentences']].values
X_test_num = X_test[['num_of_characters', 'num_of_sentences']].values

# 3. Combine TF-IDF features with numerical features
X_train_combined = hstack([X_train_tfidf, X_train_num])
X_test_combined = hstack([X_test_tfidf, X_test_num])

print('Number of feature words: ', len(vectorizer.get_feature_names_out()))
````

````python
X_train_combined.shape
````

## 4.5. Resampling

We experimented with various resampling techniques, including SMOTE, Random Over-Sampling and Random Under-Sampling. Among these, Random Over-Sampling provided the best results. 

````python
# Apply Random Over-Sampling on the vectorized data
ros = RandomOverSampler(random_state=101)
X_train_resampled, y_train_resampled = ros.fit_resample(X_train_combined, y_train)
````

````python
X_train_resampled.shape
````

# <div style="text-align:center; border-radius:30px 30px; padding:9px; color:white; margin:0; font-size:110%; font-family:Arial; background-color:#419D78; overflow:hidden"><b> 5. Model Traning and Evaluation </b></div>

````python
# Define a dictionary of classifiers with their specific parameters.
# Note: The hyperparameters for these classifiers were chosen after performing GridSearchCV to optimize performance.
classifiers = {
    'Bernoulli Naive Bayes': BernoulliNB(alpha=0.1, binarize=0.0),
    'Decision Tree': DecisionTreeClassifier(max_depth=9, min_samples_split=5, random_state=101), 
    'Logistic Regression': LogisticRegression(solver='liblinear', penalty='l1', C=10, random_state=101),
    'XGB': XGBClassifier(learning_rate=0.2, max_depth=7, n_estimators=500, random_state=101, tree_method='gpu_hist')
}
````

````python
# Initialize a list to store accuracy scores for each classifier
accuracy_scores = []

# Iterate over each classifier and its name in the classifiers dictionary
for name, clf in classifiers.items():
    clf.fit(X_train_resampled, y_train_resampled)
    y_pred = clf.predict(X_test_combined)
    accuracy = accuracy_score(y_test, y_pred)
    
    print("\n")
    print("For", name)
    print("Accuracy:", accuracy)
    
    # Compute the confusion matrix for the predictions
    # 'lbl_enc.classes_' provides the class labels for the confusion matrix and classification report
    labels = lbl_enc.classes_
    conf_matrix = confusion_matrix(y_test, y_pred)
    print(classification_report(y_test, y_pred, target_names=labels))
    
    # Plot the confusion matrix using a heatmap
    # Annotate each cell with the numeric value of the confusion matrix
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Greens', xticklabels=labels, yticklabels=labels)
    plt.xlabel('Predicted')  # Label for x-axis
    plt.ylabel('Actual')     # Label for y-axis
    plt.title(f'Confusion Matrix for {name}')  # Title for the heatmap
    plt.show()  # Display the heatmap
    
    # Append the accuracy score to the list
    accuracy_scores.append(accuracy)
````

````python
# Create a DataFrame to store classifier names and their corresponding accuracy scores
accuracies_df = pd.DataFrame({'Classifier': classifiers.keys(), 'Accuracy': accuracy_scores}).sort_values('Accuracy', ascending=False)

plt.figure(figsize=(12, 8))
palette = dict(zip(accuracies_df['Classifier'], colors[:4]))

# Create a bar plot to visualize the accuracy of each classifier
sns.barplot(x='Classifier', y='Accuracy', data=accuracies_df, palette=palette)

plt.title("Classifier Accuracy Comparison")
plt.ylim(0, 1)

plt.tight_layout()
plt.show()
````
